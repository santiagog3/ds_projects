```python
import requests
import json
import pandas as pd
import numpy as np
from datetime import date
```

```python
query = 'adam grant'
maxResults = 40          #Maximum allowable 40
```

```python
url = 'https://www.googleapis.com/books/v1/volumes'

r = requests.get(url=url,
params={'q': query,
                    'maxResults':maxResults,
                    'printType':'books'},
)
```

```python
response = json.loads(r.text)
```

```python
results = []

# Total items
print('Total items: ',(response['totalItems']))

for i in range(len(response['items'])):
    
    items = response['items'][i]
        book_id= items['id']
    title = items['volumeInfo']['title']

    try:
        subtitle= items['volumeInfo']['subtitle']
    except KeyError:
        subtitle=np.nan

    try:
        category= items['volumeInfo']['categories']
    except KeyError:
        category=np.nan

    try:
        authors=items['volumeInfo']['authors']
    except KeyError:
        authors=np.nan

    try:
        published_date=items['volumeInfo']['publishedDate']
    except KeyError:
        published_date=np.nan 
    
    try:
        averageRating = items['volumeInfo']['averageRating']
    except KeyError:
        averageRating=np.nan 


   
    try:
        description=items['volumeInfo']['description']
    except KeyError:
        description=np.nan 

# Create dataset
    results.append({
        'book_id':book_id,
        'title':title,
        'subtitle' : subtitle,
        'category':category,
        'authors':authors,
        'published_date': published_date,
        'averageRating':averageRating,
        'description':description
        }
    )


df = pd.DataFrame(results)
```
